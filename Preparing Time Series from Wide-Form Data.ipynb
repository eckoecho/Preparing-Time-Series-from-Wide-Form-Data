```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticks
import matplotlib.dates as mdates
```

```python
#Setting default sizes and color for plots
plt.rcParams['figure.facecolor']='white'
plt.rcParams['figure.figsize']=(12,4)
```

```python
df = pd.read_csv("Data/City_zhvi_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv")
df.head()
```

```python
large_cities = df.head(4).copy()
large_cities
```

```python
#Getting columns to keep
large_cities.columns
```

```python
#putting columns to keep in id_vars
#Using melt to turn from wide-form to long-form
melted = pd.melt(large_cities,id_vars=['RegionID', 'SizeRank', 'RegionName', 'RegionType', 'StateName',
       'State', 'Metro', 'CountyName'] )
melted
```

```python
#Convert the melted dates into datetime datatype
melted["Date"] = pd.to_datetime(melted["variable"])
melted
```

```python
# Make the datetime column the index
melted = melted.set_index("Date")
melted
```

```python
#Resample the dataframe as monthly frequency, grouped by City
monthly_grouped = melted.groupby('RegionName').resample('MS').mean('value')
monthly_grouped#["value"]
```

```python
unstacked.columns
```

```python
#Unstacking City Index, turns into columns
unstacked = monthly_grouped["value"].unstack(level=0)
unstacked
```

```python
def thousands(x,pos):
    """function for use wth matplotlib FuncFormatter -  formats money in millions"""
    new_x = x/1_000
    return f"{new_x:,.0f}K"
# Create the formatters
fmt_thous = mticks.FuncFormatter(thousands)
```

```python
#Plotting each city
ax = unstacked.plot()#subplots=True),# sharex=False, layout=(-1,2))
#Title for my plot
ax.set_title("Average House Values in the US");
# Reformat the y-axis ticks to use thousands of dollars with a "K" at the end
ax.yaxis.set_major_formatter(fmt_thous)
```
